### Подготовка датасета

```python
data=pd.read_stata('Счастье_RLMS.dta',convert_categoricals=False)
```

```python
# обработка пропусков
# убираем неинформативные переменные, где сплошные пропуска
# работает долго - 20 мин
for i in data.columns:
    data.loc[(data[i]> 99999996, i)] = np.nan
    if len(pd.unique(data[i]))<2:
        print(str(i)+str(pd.unique(data[i])))
        data=data.drop(labels=i,axis=1)
```

```python
# праавильно задать типы данных
data = data.apply(pd.to_numeric, errors='coerce')
print(data.shape)
```

```python
# определяем, какие из оставшихся переменных адекватны по количеству пропусков
# выбраны те, где их не больше половины
total_rows=len(data['age'])
vars_to_use=[]

print('total_rows: '+str(total_rows))

for i in data.columns:
    propuski=sum(data[i].isna())
    if propuski<total_rows/2:
        vars_to_use.append(i)
        print(str(i)+", пропусков: "+str(propuski))
```

```python
# какие есть нормальные переменные
# их количество
print(len(vars_to_use))
# их названия
print(vars_to_use)

data=data[vars_to_use]
```

```python
# интересные для исследования переменные из оставшегося
fields=['idind','year','region','status','popul', # номер индивида - год - регион - тип насел. пункта - население пункта
        'marst','OCCUP08', 'educ', 'diplom','H5', # семейный статус - занятость - образование - пол
        'age', # возраст
        'I4', # национальность
        'J1','J1_1_1', 'J1_1_2', 'J1_1_3', 'J1_1_4','J4_1', 'J6', # занятость - удовлетворённость ею - отрасль - есть подчинённые
        'J6_1A', 'J6_1B', 'J6_2','J7_1', # длительность рабочего дня (часов - мин) - рабочей недели (часов) - дней в месяце
        'J10','J60', 'J31','J60_5A', # деньги за работу - за всё - риск потери работы - изм-е материальн положения за год
        'J61','J62', 'J63', 'J64', 'J65', # ожидания (станут ли жить лучше) - самооценка своего положения и жизни
        'M20_7','L5','M3', # инвалид - проблемы со здоровьем - субъективная оценка здоровья
        'J72_18', # религиозность
        'M170' # счастье
]
```

```python
data = data[fields]
data.drop(labels=['idind', 'year','region'], axis=1).describe()
```
